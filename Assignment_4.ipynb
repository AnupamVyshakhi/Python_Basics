**1.1 STRING INDEXING**

Postive Indexing

Negative Indexing

```python
a="Anupam"
print(a[0])
print(a[5])
print(a[3])
print(a[-1])
print(a[-5])
print(a[-6])
```

**1.2 STRING SLICING**

**SYNTAX**

[start:end:step]

```python
a="Anupam"
print(a[1:5:1])
```

```python
print(a[2:4:1])
```

```python
print(a[0:5:2])
```

```python
print(a[0:5:3])
```

**1.No Step Size**

```python
a="Anupam"
print(a[1:3])
```

**2.No Step Size**

```python
a="Anupam"
print(a[1:3:])
```

**2.No Start Index**

```python
a="Anupam"
print(a[:5:2])
```

**3.No stop Index**

```python
a="Anupam"
print(a[1::2])
```

**4.No start Index No stop Index**

```python
a="Anupam"
print(a[::2])
```

**5.No Start Stop or step Index**

```python
a="Anupam"
print(a[::])
```
