remove() can be used to delete a single element from the list.

But pop() can also be used to delete an element if the index value of that element is known. If no index value is specified, pop() can be used to delete the last element of the list.
eg:

```python
new_list=[1,2,3,9,4,5,6]
new_list.remove(9)
new_list.pop()
print(new_list)
```

```python
new_list.pop(2)
print(new_list)
```
