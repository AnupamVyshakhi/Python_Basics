#**3. SWAPPING VARIABLES IN PYTHON**

```python
a=10
b=5
print(f"The value of a is {a} and b is {b}")
temp=a
a=b
b=temp
print("AFTER SWAPPING")
print(f"The values of a is {a} and b is {b}")
```

```python
a=20
b=30
print(f"The value of a is {a} and value of b is {b}")
a,b=b,a
print("AFTER SWAPPING")
print(f"The value of a is {a} and value of b is {b}")
```
