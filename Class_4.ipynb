#**3.SWAPPING VARIABLES IN PYTHON**

```python
a=10
b=5
print(f"The value of a is {a} and value of b is {b}") 
c=a
a=b
b=c
print(f"After swapping::The value of a is {a} and value of b is {b}")
```

```python
a=20
b=30
print(f"The value of a is {a} and value of b is {b}") 
a,b=b,a
print(a)
print(b)
print(f"After swapping::The value of a is {a} and value of b is {b}") 
```
