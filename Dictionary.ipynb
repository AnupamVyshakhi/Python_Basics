#**2.CREATING DICIONARY USING LIST**

```python
name=["Anupam","Aninditha","Nahida"]
mark=[90,91,67]
a2=dict(zip(name,mark))
print(a2)
print(type(a2))
```

#**3.ADD VALUES**

```python
a2
```

```python
a2["Nandana"]=70
print(a2)
```

#**4.UPDATE**

```python
a2.update({"Ashish":32,"Akash":45})
print(a2)
```

#**5.DELETE VALUES**

```python
del a2['Nandana']
print(a2)
```

```python
print(a2)
```

```python
g=a2.pop("Nahida")
print(g)
print(a2)
```
